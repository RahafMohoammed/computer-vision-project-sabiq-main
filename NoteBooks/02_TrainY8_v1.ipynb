```python
import kagglehub

path = kagglehub.dataset_download("aliabdelmenam/rdd-2022")
print("Path:", path)
```

```python
import os

DATASET_PATH = path
for item in os.listdir(path):
    candidate = os.path.join(path, item)
    if os.path.isdir(candidate) and os.path.exists(os.path.join(candidate, 'train')):
        DATASET_PATH = candidate
        break

for split in ['train', 'val', 'test']:
    imgs = len(os.listdir(os.path.join(DATASET_PATH, split, 'images')))
    print(f"  {split}: {imgs} images")
```

```python
import os

DATASET_PATH = path
for item in os.listdir(path):
    candidate = os.path.join(path, item)
    if os.path.isdir(candidate) and os.path.exists(os.path.join(candidate, 'train')):
        DATASET_PATH = candidate
        break

for split in ['train', 'val', 'test']:
    imgs = len(os.listdir(os.path.join(DATASET_PATH, split, 'images')))
    print(f"  {split}: {imgs} images")

yaml_content = f"""path: {DATASET_PATH}
train: train/images
val: val/images
test: test/images

nc: 5
names:
  0: D00
  1: D10
  2: D20
  3: Other
  4: D40
"""
YAML_PATH = "/content/data.yaml"
with open(YAML_PATH, 'w') as f:
    f.write(yaml_content)

print(f"data.yaml saved at {YAML_PATH}")
```

```python
from ultralytics import YOLO

model = YOLO('yolov8m.pt')

results = model.train(
    data="/content/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    patience=10,
    device=0,
    name="sabiq_v1",
    save=True,
    plots=True,
    verbose=True,
    workers=4,
    cache='ram',
)

print("\nTraining complete!")
```
